```python
import pandas as pd
import requests         # grab web-page
from bs4 import BeautifulSoup as Soup  # parse web-page
import datetime         # format date/time
from tqdm.auto import tqdm
```

```python
jobs = []
for j in tqdm(range(0,200,10), desc="Getting Job Links..."): # calling entries
    position, location='data analyst','dublin'
    resp = requests.get(f'https://ie.indeed.com/jobs?q={position.replace(" ", "+")}&l={location.replace(" ", "+")}&start={str(j)}')
    soup = Soup(resp.text, 'html.parser')
    joblinks = soup.findAll("a", {"class":"tapItem"})
    for joblink in joblinks:
        jobs.append(f"https://ie.indeed.com{joblink['href']}")

    
```

```python
res = []
for job in tqdm(jobs, desc="Getting information from links..."):
    resp = requests.get(job)
    soup = Soup(resp.text)
    try:
        jobTitle = soup.find("h1", {"class":"jobsearch-JobInfoHeader-title"}).text
    except:
        jobTitle = ""
    try:
        jobCompany = soup.find("div", {"class":"icl-u-lg-mr--sm icl-u-xs-mr--xs"}).text
    except:
        jobCompany = ""
    try:
        jobLocation = soup.find("div", {"class":"icl-u-xs-mt--xs icl-u-textColor--secondary jobsearch-JobInfoHeader-subtitle jobsearch-DesktopStickyContainer-subtitle"}).findAll("div")[-1].text
    except:
        jobLocation = ""
    try:
        jobDescription = soup.find("div", {"class":"jobsearch-jobDescriptionText"}).text
    except:
        jobDescription = ""
    
    res.append({"Job Title":jobTitle, "Company":jobCompany, "Location":jobLocation, "Description":jobDescription.replace("\n", " ")})
```

```python
df = pd.DataFrame(res)
df.to_csv("output.csv", index=False)
```
